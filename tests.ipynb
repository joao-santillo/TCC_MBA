```python
import matplotlib.pyplot as plt

# Dados de distribuição de classes
classes = ['Normal', 'Tumor Benigno', 'Tumor Maligno']
quantidades = [133, 437, 210]

# Criação do gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(classes, quantidades, color=['blue', 'green', 'red'])
plt.xlabel('Classes')
plt.ylabel('Quantidade de Imagens')
plt.title('Distribuição de Classes no Conjunto de Dados')
plt.show()
```

```python


# Dados de distribuição de classes
classes = ['Normal', 'Tumor Benigno', 'Tumor Maligno']
quantidades = [133, 437, 210]
total = sum(quantidades)

# Cálculo das porcentagens
percentuais = [(q / total) * 100 for q in quantidades]

# Criação do gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(classes, percentuais, color=['blue', 'green', 'red'])
plt.xlabel('Classes')
plt.ylabel('Porcentagem de Imagens (%)')
plt.title('Distribuição Percentual de Classes no Conjunto de Dados')
plt.ylim(0, 70)  # Definindo o limite do eixo y para melhor visualização
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados dos três cenários
metricas = ['Acurácia', 'Precisão', 'F1-Score', 'Revocação']
cenario_1 = [0.7886, 0.7870, 0.7838, 0.7834]
cenario_2 = [0.8000, 0.8252, 0.8169, 0.8153]
cenario_3 = [0.7990, 0.8208, 0.8112, 0.8089]

# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(metricas, cenario_1, marker='o', label='Cenário 1')
plt.plot(metricas, cenario_2, marker='o', label='Cenário 2')
plt.plot(metricas, cenario_3, marker='o', label='Cenário 3')

# Adicionando título e rótulos aos eixos
plt.title('Comparação das Métricas de Avaliação nos Três Cenários')
plt.xlabel('Métricas')
plt.ylabel('Valores')
plt.ylim(0.75, 0.85)

# Adicionando legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
